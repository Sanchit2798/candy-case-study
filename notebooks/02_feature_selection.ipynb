```python
import sys
import pandas as pd
from sklearn.linear_model import LinearRegression
sys.path.append("../") 

# Importing the functions from src
from src.ml.recursive_feature_elimination import RecursiveFeatureElimination
```

```python
df  = pd.read_csv('../data/processed/candy-data-processed.csv')
print('data shape', df.shape)
df.head()
```

```python
# Categorize the variables
Target          = "WinPercent"
FlavorFeats     = ["Chocolate", "Fruity", "Caramel", "PeanutyAlmondy", "Nougat", "CrispedRiceWafer"]
FormFeats       = ["Hard", "Bar", "PluriBus"]
ContinuousFeats = ["SugarPercent", "PricePercent"]
Features        = FlavorFeats + FormFeats + ContinuousFeats
```

## RFE (Recursive Feature Elimination with Linear Regression)

```python
rfe = RecursiveFeatureElimination(LinearRegression(),
                df, Features, 
                [],
                Target, min_features_num = 1,
                standardisation = True,
                test_size = 0.1,
                optmise_hyperparameters_for_models = True,
                score = 'R2 score',
                thresh = None,
                verbose=True)
(features_for_each_iteration, scores_for_each_iteration, feature_removed_for_each_iteration, 
score_list_with_each_feature_removed_for_each_iteration) = rfe.run(5)
```

```python
results_recursive_elimination_df = pd.DataFrame({"features_for_each_iteration" : features_for_each_iteration,
            "scores_for_each_iteration" : scores_for_each_iteration, "removed_feature":feature_removed_for_each_iteration})
results_recursive_elimination_df
```
